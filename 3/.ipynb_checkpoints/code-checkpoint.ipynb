```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0, 0, 0, 0, 0]
p = [1, 2, 3, 4, 5]
k = 0.1
m = 2
```

```python
plt.figure(figsize=(20, 10))
plt.title(f"Momentum vs X")
plt.xlabel("X")
plt.ylabel("Momentum")

X_mean = []
for index, _x in enumerate(x):
    X = [_x]
    P = [p[index]]
    H = (0.5 * k * _x ** 2) + P[-1] ** 2 / (2 * m)
    for timestamp in range(1000):
        X.append(X[-1] + P[-1] * 0.1 / m)
        P.append(P[-1] - X[-1] * 0.1 * k)
    
    plt.plot(X, P, label = f"Hamiltonian: {H}")
    X_mean.append([X, H])

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.title(f"Mean Square Displacement vs Time")
plt.xlabel("Time")
plt.ylabel("MSD")

for X_H in X_mean:
    X = np.array(X_H[0])
    H = X_H[1]
    plt.plot(X ** 2, label = f"H: {H}")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
import statistics
```

# $I_{1}$

```python
area_bound = 1 * 3
I1_vs_n = []
I1_actual_vs_n = []

for n in range(1, 1000):
    hit = 0
    for point in range(n):
        x = random.random()
        y = random.uniform(0, 3)
        if y <= 3 * x**2:
            hit += 1
    area_under = (hit / n) * area_bound
    I1_vs_n.append(area_under)
    I1_actual_vs_n.append(1)
    
plt.title("Area vs N for I1")
plt.xlabel("N")
plt.ylabel("Area")
plt.plot(I1_vs_n, label = "Simulation")
plt.plot(I1_actual_vs_n, label = "Actual")
plt.legend()
plt.show()
```

## Standard Deviation

```python
area_bound = 1 * 3
I1_vs_n = []
I1_actual_vs_n = []
N = 1000 # change for N = 20 and 1000
for trial in range(100):
    hit = 0
    for point in range(N):
        x = random.random()
        y = random.uniform(0, 3)
        if y <= 3 * x**2:
            hit += 1
    area_under = (hit / N) * area_bound
    I1_vs_n.append(area_under)
    I1_actual_vs_n.append(1)

print("Standard Deviation: ", statistics.stdev(I1_vs_n))
    
plt.title(f"Integral 1 vs Trial N={N}")
plt.xlabel("Trial")
plt.ylabel("Integral")
plt.plot(I1_vs_n, label = "Simulation")
plt.plot(I1_actual_vs_n, label = "Actual")
plt.legend()
plt.show()
```

## Standard Deviation Graph vs N

```python
area_bound = 1 * 3
Std1_vs_n = []
N = 1000
for points in range(1, N):
    I1_vs_points = []
    for trial in range(100):
        hit = 0
        for point in range(points):
            x = random.random()
            y = random.uniform(0, 3)
            if y <= 3 * x**2:
                hit += 1
        area_under = (hit / points) * area_bound
        I1_vs_points.append(area_under)

    Std1_vs_n.append(statistics.stdev(I1_vs_points))

plt.title("Std Deviation 1 vs 100 Trials vs N")
plt.xlabel("N")
plt.ylabel("Std Deivation")
plt.plot(Std1_vs_n, label = "Simulation")
plt.show()
```

# $I_{2}$

```python
volume_bound = 1 * 1 * 1
I2_vs_n = []
I2_actual_vs_n = []
for n in range(1, 1000):
    hit = 0
    for point in range(n):
        x = random.random()
        y = random.random()
        z = random.random()
        if z <= y * x**2:
            hit += 1
    volume_under = (hit / n) * volume_bound
    I2_vs_n.append(volume_under)
    I2_actual_vs_n.append(1 / 6)
    
plt.title("Volume vs N for I2")
plt.xlabel("N")
plt.ylabel("Volume")
plt.plot(I2_vs_n, label="Simulation")
plt.plot(I2_actual_vs_n, label="Actual")
plt.legend()
plt.show()
```

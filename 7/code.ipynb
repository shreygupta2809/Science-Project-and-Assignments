```python
import numpy as np
import random
import matplotlib.pyplot as plt
import math
```

```python
N = 100
initial_spin_array = np.array([[-1, 1][random.randint(0, 1)] for i in range(N)])
```

```python
J = 2
def energy_model(s_array):
    energy = 0.0
    for i in range(len(s_array)):
        energy += -J * s_array[i] * s_array[(i + 1) % len(s_array)]
    return energy
```

```python
M_vs_T = []
H_vs_T = []
for temperature in range(1, 100):
    spin_array = initial_spin_array.copy()
    e_sum = 0
    m_sum = 0
    e_old = energy_model(list(spin_array))
    for i in range(1000):
        temp_spin_array = spin_array.copy()
        random_index = np.random.random(temp_spin_array.shape) < 0.2
        temp_spin_array[random_index] = -temp_spin_array[random_index]
        e_new = energy_model(list(temp_spin_array))
        if e_new < e_old:
            e_old = e_new
            spin_array = temp_spin_array.copy()
        else:
            is_accept = math.exp((e_old - e_new) / temperature)
            if is_accept >= random.random():
                e_old = e_new
                spin_array = temp_spin_array.copy()
        e_sum += e_old
        m_sum += np.sum(spin_array) / spin_array.shape[0]
        
    M_vs_T.append(m_sum / 1000)
    H_vs_T.append(e_sum / 1000)
    

plt.title("Energy vs Temperature")
plt.xlabel("Temperature")
plt.ylabel("Energy")
plt.plot(H_vs_T)
plt.show()

plt.title("Magnetic Moment vs Temperature")
plt.xlabel("Temperature")
plt.ylabel("Magnetic Moment")
plt.plot(M_vs_T)
plt.show()
```

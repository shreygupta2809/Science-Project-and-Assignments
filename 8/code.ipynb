```python
import matplotlib.pyplot as plt
import math
import random
```

```python
r = 1.1
alpha = 0.4
c = 0.4
beta = 1
k = 2
n_1 = [10, 12, 15, 0]
n_2 = [0, 2, 4, 10]
dT = 0.01
```

```python
for index, _n in enumerate(n_1):
    N_1 = [_n]
    N_2 = [n_2[index]]
    for timestamp in range(10000):
        N_1.append(N_1[-1] + dT * (r * N_1[-1] * (1 - N_1[-1] / k) - alpha * N_1[-1] * N_2[-1]))
        N_2.append(N_2[-1] + dT * (-c * N_2[-1] + beta * N_1[-1] * N_2[-1]))
    plt.title(f"Population vs Time, Prey: {_n}, Predator: {n_2[index]}")
    plt.xlabel("Time")
    plt.ylabel("Population")
#     plt.plot(N_1, N_2, label = f"N1: {_n}, N2: {n_2[index]}")
    plt.plot(N_1, label = "Prey N1")
    plt.plot(N_2, label = "Predator N2")
    plt.legend()
    plt.show()
```
